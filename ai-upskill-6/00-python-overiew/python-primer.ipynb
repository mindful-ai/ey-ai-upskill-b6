# 3. Looping

```python
i = 1
print("5 X ", i, " = ", 5*i)
```

```python
for i in range(1, 11):
    print("5 X ", i, " = ", 5*i)
```

```python
i = 1
while i < 11:
    print("5 X ", i, " = ", 5*i)
    i += 1
```

```python
for i in range(1, 11):
    print("5 X ", i, " = ", 5*i)
    if i == 5:
        break
```

```python
for i in range(1, 11):
    if i == 5:
        continue
    print("5 X ", i, " = ", 5*i)
```

# 4. Problem Solving

##### Write a program to determine if a number is prime or not

```python
n = int(input("Enter a number: "))
prime = True
for i in range(2, n):
    if n % i == 0:
        prime = False
        break
if prime == True:
    print("The number is prime")
else:
    print("Then number is not prime")
```

# 5. Functions

##### Extract all the prime numbers between a use given range

```python
def checkprime(n):
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
```

```python
checkprime(13)
```

```python
checkprime(34)
```

```python
m = int(input("Enter start: "))
n = int(input("Enter end: "))
if (m < n):
    primes = []
    for n in range(m, n+1):
        if(checkprime(n)):
            primes.append(n)
primes
```
